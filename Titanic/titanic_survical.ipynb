```python
# Numerical Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median"))
    , ("numscaler", StandardScaler())
])

```

```python
# Categorical Pipeline
from sklearn. preprocessing import OneHotEncoder, OrdinalEncoder

cat_pipeline = Pipeline([
    ("cat_ordinal", OrdinalEncoder())
    , ("cat_imputer", SimpleImputer(strategy="most_frequent"))
    , ("cat_onehot", OneHotEncoder(sparse_output=False))
])
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = ["Age", "SibSp", "Parch", "Fare"]
cat_attribs = ["Pclass", "Sex", "Embarked"]


column_transformer = ColumnTransformer([
    ("num_col_transform", num_pipeline, num_attribs)
    , ("cat_col_transform", cat_pipeline, cat_attribs)
])
```

```python
X_train = column_transformer.fit_transform(train_data)
```

```python
# Now train the first model
y_train = train_data["Survived"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)

random_score = cross_val_score(rnd_clf, X_train, y_train, cv=10)
```

```python
random_score.mean()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier(n_neighbors= 4, weights='distance')

neighbour_score = cross_val_score(knn_clf, X_train, y_train, cv=10)
```

```python
neighbour_score.mean()
```

```python
from sklearn.svm import SVC

svm_clf = SVC(gamma="auto")

svm_score = cross_val_score(svm_clf, X_train, y_train, cv=10, scoring="accuracy")
```

```python
svm_score.mean()
```

```python
# SVC wohl der beste Estimator

train_data_frame = train_data.drop('Survived', axis=1)
train_labels = train_data['Survived'].copy()

svm_clf_with_preprocessor = Pipeline([
    ("preprocessor", column_transformer)
    , ("scf_classifier", SVC())
])

#svm_clf_with_preprocessor.fit(train_data_frame, train_labels)
```

```python
# Optimierung mittels einer Grid Search
from sklearn.model_selection import GridSearchCV

param_grid = {
    'scf_classifier__C': [1, 10],                    # Regularization parameter
    'scf_classifier__kernel': ['linear', 'rbf'],       # Kernel type
    'scf_classifier__gamma': ['scale', 0.1, 0.01],  # Kernel coefficient for 'rbf', 'poly', and 'sigmoid'
    
}

grid_search_svc = GridSearchCV(svm_clf_with_preprocessor, param_grid=param_grid, cv=5, scoring="accuracy")

grid_search_svc.fit(train_data_frame, train_labels)
```

```python
# Old Model: 0.8249313358302123

best_estimator = grid_search_svc.best_estimator_
best_parameters = grid_search_svc.best_params_
best_scroe = grid_search_svc.best_score_
```

```python
print("Best Score: ", best_scroe)
```
