```python
import numpy as np
```

```python
print(zeroArray)
```

```python
a = np.zeros((4,2))
print(a)
```

```python
print(a.shape)  # Beschreibung der Achsen 
print(a.ndim)   # Anzahl der Array Dimensionen
print(a.size)   # Anzahld der gesamten Elemente
print(a.dtype)  # Datatype
```

```python
# Uninitialized array --> Whatever lies in memory will be written inside
empty_arr = np.empty((4,3))
print(empty_arr)

full_arr = np.full((5,3), 8.4)
print(full_arr)
print(full_arr.dtype)
```

```python
# From Python arrays
from_python_arr = np.array([[1, 2, 3, 4], [11, 22, 33, 44]])
print(from_python_arr)
```

```python
from_range_arr = np.arange(1,7, 0.4)
print(from_range_arr)
print(from_range_arr.dtype)
print(from_range_arr.shape)
```

```python
# Random Values
np.random.seed(56)

random_array = np.random.rand(4,5)

print(random_array)
print(random_array.dtype)
```

```python
# Size in Bytes of each item
print(random_array.itemsize)
```

```python
# Reshaping
# The size must remain the same

range_arr = np.arange(24)
print(range_arr)
print(range_arr.size)
print(range_arr.ndim)
```

```python
range_arr.shape = (6,4)
print(range_arr)
print(range_arr.size)
print(range_arr.ndim)
```

```python
# Item access
range_arr[1,2] = 999
print(range_arr)
```

```python
# Alle Artithmetischen Operationen funktionieren auch hier
arrit_a = np.array([14, 23, 32, 41])
arrit_b = np.array([5,  4,  3,  2])

sum_a_b = arrit_a + arrit_b
print(sum_a_b)

diff_a_b = arrit_a - arrit_b
print(diff_a_b)

mult_a_b = arrit_a * arrit_b
print(mult_a_b)
```

```python

h = np.arange(5).reshape(1, 1, 5)

print(h)

h_added = h + [10, 20, 30, 40, 50] # same as: h + [[[10, 20, 30, 40, 50]]]

print(h_added)

k = np.arange(4).reshape(2,2)
print(k)
k_added_dummys = k + [100, 200] # same as h + [[100, 200], [100, 200]]
print(k_added_dummys)
```

```python
# Conditionals apply element wise
test_conditional_arr = np.arange(10).reshape((5,2))
print(test_conditional_arr)

print("\n")
greater_4_arr = test_conditional_arr > 4
print(greater_4_arr)
print(greater_4_arr.shape)

print("\n")
only_greater_4 = test_conditional_arr[test_conditional_arr > 3]
print(only_greater_4)
only_greater_reshaped = only_greater_4.reshape(2,3)
print(only_greater_reshaped)

print("\n")
only_dividable_by_two = test_conditional_arr[test_conditional_arr % 2 == 0]
print(only_dividable_by_two)
print(only_dividable_by_two.size)
```
