```python
# Alle Artithmetischen Operationen funktionieren auch hier
arrit_a = np.array([14, 23, 32, 41])
arrit_b = np.array([5,  4,  3,  2])

sum_a_b = arrit_a + arrit_b
print(sum_a_b)

diff_a_b = arrit_a - arrit_b
print(diff_a_b)

mult_a_b = arrit_a * arrit_b
print(mult_a_b)
```

```python

h = np.arange(5).reshape(1, 1, 5)

print(h)

h_added = h + [10, 20, 30, 40, 50] # same as: h + [[[10, 20, 30, 40, 50]]]

print(h_added)

k = np.arange(4).reshape(2,2)
print(k)
k_added_dummys = k + [100, 200] # same as h + [[100, 200], [100, 200]]
print(k_added_dummys)
```

```python
# Conditionals apply element wise
test_conditional_arr = np.arange(10).reshape((5,2))
print(test_conditional_arr)

print("\n")
greater_4_arr = test_conditional_arr > 4
print(greater_4_arr)
print(greater_4_arr.shape)

print("\n")
only_greater_4 = test_conditional_arr[test_conditional_arr > 3]
print(only_greater_4)
only_greater_reshaped = only_greater_4.reshape(2,3)
print(only_greater_reshaped)

print("\n")
only_dividable_by_two = test_conditional_arr[test_conditional_arr % 2 == 0]
print(only_dividable_by_two)
print(only_dividable_by_two.size)
```

```python
a = np.array([[2,6,7],[8,5,2]])
print(a)
print(a.shape)

# Methods calculate over all elements regardless of the array shape
print("mean=", a.mean())
print("max=", a.max())
print("min=", a.min())
print("sum=", a.sum())
```

```python
# universal functions (static in np lib. No Object Method like above)

print(b)

for func in (np.abs, np.sqrt, np.exp, np.log, np.sign, np.ceil, np.modf, np.isnan, np.cos):
    # Example: np.square(b) --> Returns a new np Array
    print(f"\n{func.__name__}")
    print(func(b))
```

```python
# multi dimensional index
b_multi = np.arange(48).reshape(4, 12)

print(b_multi)
```

```python
print(b_multi[1, 3])
print(b_multi[2, :]) # row 3, all columns
print(b_multi[:, 2]) # colum 3, all rows
print(b_multi[0:2, 2:5]) # rows 0 to 2, columns with index 2 to 4
print(b_multi[(0, 2), 2:5]) # rows 0 AND 2, columns with index 2 to 4
```

```python
# Solving scalar equotation
# x is actually x1 and y is actually x2
# so this is basically Ax = b
# A [[2, 6], [5, 3]] --> As vector
# b [6, -9] --> as Vector
# x [x, y] --> as Vectory
# So now lets solve this and calculate x, y with numpy
#2x+6y=6 
#5x+3y=âˆ’9
from numpy import linalg

constants = np.array([[2, 6], [5, 3]])
results = np.array([6, -9])

solution = linalg.solve(constants, results)
```
